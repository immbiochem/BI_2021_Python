#### Класс - лимфоцит
Если не указано при создании, генерируется рандомный клеточный рецептор, либо переводится в строку значение, заданное пользователем.

Атрибуты: рецептор, лимитер делений, статус (Жив/Мертв).

При создании дается предел Хейфлика для делений (52 деления).

Есть функция делиться - возвращается обект класса Lymphocyte, с таким же рецептором, как и у родителя, лимит делений уменьшается у обоих на 1.

Есть функция дать антиген - вызывает деление, если предоставленный антиген совпадает, если не совпадает - убивает клетку (две разные функции учитывают неспецифическую и специфическую пролиферацию лимфоцитов).

При обнулении предела Хейфлика лимфоцит не умирает, но делиться уже не может. 

```python
import string
import numpy as np
```

```python
class Lymphocyte:
    def __init__(self, receptor = '0'):
        if receptor == '0':
            self.receptor = ''.join([i for i in np.random.choice(list(string.ascii_letters), size=3)])
        else:
            self.receptor = str(receptor)
        
    hayflick_limit = 52    
    status = 'Alive'
    
    def proliferate(self):
        if self.status == 'Alive' and self.hayflick_limit>0:
            dauther_cell = Lymphocyte(self.receptor) 
            self.hayflick_limit -= 1
            dauther_cell.hayflick_limit = self.hayflick_limit
            return dauther_cell
        else:
            return "Wasted"
        
    def given_antigen(self, antigen):
        if antigen == self.receptor:
            return self.proliferate()
        else:
            self.status = 'Dead'
```

#### Демонстрация класса

```python
a = Lymphocyte()
b = Lymphocyte()
print("a: ", f"Receptor: {a.receptor} | Limit: {a.hayflick_limit} | Status: {a.status}")
print("b: ", f"Receptor: {b.receptor} | Limit: {b.hayflick_limit} | Status: {b.status}")
```

```python
c = a.proliferate()
print("a: ", f"Receptor: {a.receptor} | Limit: {a.hayflick_limit} | Status: {a.status}")
print("c: ", f"Receptor: {c.receptor} | Limit: {c.hayflick_limit} | Status: {c.status}") 
```

```python
d = b.given_antigen(b.receptor)
print("b: ", f"Receptor: {b.receptor} | Limit: {b.hayflick_limit} | Status: {b.status}")
print("d: ", f"Receptor: {d.receptor} | Limit: {d.hayflick_limit} | Status: {d.status}")
```

```python
a.given_antigen(b.receptor)
print("a: ", f"Receptor: {a.receptor} | Limit: {a.hayflick_limit} | Status: {a.status}")
a.proliferate()
```

```python
for _ in range(b.hayflick_limit):
    b.proliferate()
print("b: ", f"Receptor: {b.receptor} | Limit: {b.hayflick_limit} | Status: {b.status}")
print(b.proliferate())
print("d: ", f"Receptor: {d.receptor} | Limit: {d.hayflick_limit} | Status: {d.status}")
```

### Задание 2
Напишите класс, описывающий РНК, у него должны быть:

- Конструктор - принимает последовательность РНК и создаёт объект с ней
- Метод трансляции - возвращает строку, соответствующую белку из этой РНК по стандартному коду
- Метод обратной транскрипции - возвращает строку, соответствующую ДНК из этой РНК
- Подумайте, какие атрибуты должны быть заданы для объекта и класса

```python
from Bio.Seq import Seq
import re
```

```python
class RNA:
    def __init__(self, seq):
        pattern = re.compile('[^AUCGaugc]+')
        if not len(re.findall(pattern, seq)):
            self.sequence = Seq(seq)
        else:
            raise TypeError("This string isn't RNA")
    
    def __len__(self):
        return len(self.sequence)
        
    def translation(self):
        return str(self.sequence.translate())
    
    def reverse_transcription(self):
        return str(self.sequence.back_transcribe())
        
    
```

#### Демомстрация класса

```python
a = RNA('AGCCGU')
print(a.reverse_transcription())
print(a.translation())
print(len(a))
```

### Задание 3
Напишите класс, унаследовавшись от сэтов, который будет содержать в себе только положительные числа при создании и не будет добавлять неположительные элементы (подсказка - методы конструктора и add)

```python
class Cathode(set):
    def __init__(self, *args):
        if len(list(filter(lambda x: x < 0, args))) > 0:
            raise TypeError("Negative charge found")
        else:
            self.update(args)
    
    def add(self, x):
        if x < 0:
            raise TypeError("Negative charge found")
        else:
            self.update([x])
    
```

#### Демонстрация класса

```python
try:
    a = Cathode(-1, 2, 3)
    print('1) ', a)
except TypeError:
    print('1) Check')
try:
    b = Cathode(1, 2, 3)
    print('2)', b)
except TypeError:
    print('2) Check')    
```

```python
try:
    b.add(-1)
    print(b)
except TypeError:
    print('1) Check')
    
try:
    b.add(150)
    print(b)
except TypeError:
    print('2) Check')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from Bio import SeqIO
from Bio.SeqUtils import GC
from itertools import product

class FastQC:
    
    def __init__(self, path):
        self.path = path
        self.fastq_dict = dict()
        self.key_list = list()
        for record in SeqIO.parse(path, "fastq"):
            self.fastq_dict[record.id] = record.seq
            self.key_list.append(record.id)
        self.amount_of_reads = len(self.key_list)
    
    def seq_num(self):
        return self.amount_of_reads
    
    def len_hist(self):
        sns.histplot(list(map(lambda x: len(self.fastq_dict[x]), self.key_list)), bins=20)
        plt.grid()
        plt.show()
    
    def gc_reads(self):
        sns.histplot(list(map(lambda x: GC(self.fastq_dict[x]), self.key_list)), bins=20)
        plt.grid()
        plt.show()
    
    def gc_mean(self):
        return round(np.array(list(map(lambda x: GC(self.fastq_dict[x]), self.key_list))).mean(), 1)
    
    def four_mer_hist(self):
        
        print("Initiation")
        
        alphabet = ["A", "T", "G", "C"]
        pro_list = list(product(alphabet, repeat=4))
        list_of_four_mers = list(map(lambda x: ''.join(x), pro_list))
        array_for_fastq = np.zeros([self.amount_of_reads, len(list_of_four_mers)])
        
        print("Preparation's done")
        
        for row in range(array_for_fastq.shape[0]):
            array_for_fastq[row] = list(map(lambda x: self.fastq_dict[self.key_list[row]].count(x), list_of_four_mers))
        vector = array_for_fastq.mean(0)
        
        print("Bulid the graph")
        
        fig_dims = (20, 10)
        plt.subplots(figsize=fig_dims)
        aa = sns.lineplot(list_of_four_mers, vector, color = "red")
        aa.set(xlabel='4-mers', ylabel='Frequency, %')
        plt.xticks(rotation=90)
        plt.grid()
        plt.show()
        
    def __str__(self):
        return self.path
    
    def mean_n_count(self):
        return round(np.array(list(map(lambda x: self.fastq_dict[x].count('N'), self.key_list))).mean(), 1)
    
    def percent_of(self, k):
        value = 0
        for val in self.key_list:
            if self.fastq_dict[val].count(k) > 0:
                value += 1
        return round(value/self.amount_of_reads, 1)*100
```

#### Демонстрация класса

```python
amp = FastQC(path)
```

```python
print("Amount of reads: ", amp.seq_num())
print("List of keys: ", amp.key_list[:10])
print("Seq №1: ", amp.fastq_dict[amp.key_list[0]])
print("Mean of GC: ", amp.gc_mean())
print(amp)
```
